## 13

```python
eq_temp = italy_earthquakes
```

```python
eq_temp["Pre"] = eq_temp.Magnitude.diff().round(1).fillna(0)
```

```python
eq_temp
```

```python
eq_temp["Post"] = eq_temp.Magnitude.diff(-1).round(1).fillna(0)
```

```python
eq_temp
```

```python
more_eq = eq_temp.loc[(eq_temp.Pre >= 2) & (eq_temp.Post >= 2)] 
```

```python
more_eq
```
