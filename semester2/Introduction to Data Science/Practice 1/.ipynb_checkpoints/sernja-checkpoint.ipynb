## 23

```python
df_drop_medal = medals.drop(columns=['Summer', 'Winter'])
sum_medal = df_drop_medal.sum(axis=1)
df_drop_medal = df_drop_medal.drop(columns=['1S','2S','3S','1W','2W','3W'])
df_drop_medal.insert(1,'sum_medal',sum_medal)
df_drop_medal
```

## 24

```python
sum_medal_winter = medals['1W'] + medals['2W'] + medals['3W']
sum_medal_summer = medals['1S'] + medals['2S'] + medals['3S']
df_drop_medal.insert(2,'sum_medal_winter',sum_medal_winter)
df_drop_medal.insert(3,'sum_medal_summer',sum_medal_summer)
```

```python
df_drop_medal.loc[df_drop_medal['sum_medal_summer'] > df_drop_medal['sum_medal_winter'],:]
```

## 28

```python
score_summer = medals['1S']*3 + medals['2S']*2 + medals['3S']
score_winter = medals['1W']*3 + medals['2W']*2 + medals['3W']
score = score_summer + score_winter
score
```

```python
medals.insert(9,'score',score)
medals
```

```python
medals.sort_values(by = 'score', ascending=False)[:10]
```

## 29

```python
perGame = (score_summer/medals['Summer']).fillna(0) + (score_winter/medals['Winter']).fillna(0)
perGame
```

```python
medals.insert(10,'perGame',perGame)
medals
```

```python
medals.sort_values(by='perGame', ascending=True)[:10]
```
